# Моделирование случайных величин (СВ) при помощи библиотеки NumPy

### Импорт требуемых библиотек

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Равномерное распределение
### Гистограмма распределения

```python
# Случайная величина распределена равномерно на интервале (15, 15+10)
# Вектор x содержит 500 измерений
x = 15 + 10 * np.random.random(500)
# На графике представлена гистограмма распределения случайной величины
n, bins, patches = plt.hist(x, bins=50, normed=1, facecolor='g', alpha=0.75)
plt.grid(True)
plt.show()
```

### Закон распределения смоделированной СВ

```python
# Закон распределения F(x) = P(x < X)
# Чтобы построить закон распределения для смоделированной СВ
# выполним сортировку
x_sort = x[:]
x_sort.sort()  # вектор x_sort содержит смоделированные значения СВ,
               # отсортированные по возрастанию
F = [x/500 for x in range(500)]  # вектор v содержит значение функции распределения
                                 # для каждого наблюдения из x_sort
plt.plot(x_sort, F)
plt.axis([14, 26, -0.2, 1.2])
plt.grid(True)
plt.show()
```

### Плотность распределения смоделированной СВ

```python
# Плотность распределения f(x) = F'(x)
# Численно плотность распределения можно вычислить как приращение функции F(x),
# то есть значение f(x) = dF / dx = (F(x[i+1]) - F(x[i])) / (x[i+1] - x[i])
# Чтобы сгладить результат возьмём среднее по 10 элементам
f = [(F[i+10] - F[i])/(x_sort[i+10] - x_sort[i]) for i in range(490)]
plt.plot(x_sort[5:-5], f)
plt.axis([14, 26, min(f)-0.2, max(f)+0.2])
plt.grid(True)
plt.show()
```

## Нормальное распределение
### Гистограмма распределения

```python
# Случайная величина распределена нормально со средним значением 0 и СКО = 10
# Вектор x содержит N измерений
N = 10000
mu, sigma = 0, 15
x = mu + sigma * np.random.randn(N)
# На графике представлена гистограмма распределения случайной величины
n, bins, patches = plt.hist(x, bins=50, normed=1, facecolor='g', alpha=0.75)
plt.grid(True)
plt.show()
```

### Закон распределения смоделированной СВ

```python
# Закон распределения F(x) = P(x < X)
# Чтобы построить закон распределения для смоделированной СВ
# выполним сортировку
x_sort = x[:]
x_sort.sort()  # вектор x_sort содержит смоделированные значения СВ,
               # отсортированные по возрастанию
F = [x/N for x in range(N)]  # вектор v содержит значение функции распределения
                                 # для каждого наблюдения из x_sort
plt.plot(x_sort, F)
plt.axis([-50, 50, -0.2, 1.2])
plt.grid(True)
plt.show()
```

### Плотность распределения смоделированной СВ

```python
# Плотность распределения f(x) = F'(x)
# Численно плотность распределения можно вычислить как приращение функции F(x),
# то есть значение f(x) = dF / dx = (F(x[i+1]) - F(x[i])) / (x[i+1] - x[i])
# Чтобы сгладить результат возьмём среднее по 10 элементам
f = [(F[i+20] - F[i])/(x_sort[i+20] - x_sort[i]) for i in range(N-20)]
plt.plot(x_sort[10:-10], f)
plt.axis([-50, 50, min(f)-0.2, max(f)+0.2])
plt.grid(True)
plt.show()
```
